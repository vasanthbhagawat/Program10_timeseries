```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create date range for the month of January 2022
date_range = pd.date_range(start='2022-01-01', end='2022-01-31', freq='D')

# Create array of random temperatures for each day
temps = np.random.normal(25, 5, len(date_range))

# Create Pandas DataFrame
df = pd.DataFrame({'Temperature': temps}, index=date_range)

# Resample to weekly frequency and compute the mean
weekly_mean = df.resample('W').mean()

# Compute rolling 7-day mean
rolling_mean = df.rolling(window=7).mean()

# Compute difference between consecutive values
diff = df.diff()

# Plot the results
plt.figure(figsize=(20, 8))

plt.subplot(2, 2, 1)
plt.plot(df.index, df['Temperature'], label='Temperature')
#plt.legend()
plt.plot()
plt.title('Daily Temperature')

plt.subplot(2, 2, 2)
plt.plot(weekly_mean.index, weekly_mean['Temperature'], label='Temperature')
plt.legend()
plt.title('Weekly Mean Temperature')

plt.subplot(2, 2, 3)
plt.plot(rolling_mean.index, rolling_mean['Temperature'], label='Temperature')
plt.legend()
plt.title('7-day Rolling Mean Temperature')

plt.subplot(2, 2, 4)
plt.plot(diff.index, diff['Temperature'], label='Temperature')
plt.legend()
plt.title('Daily Temperature Difference')

plt.tight_layout()
plt.show()
```
